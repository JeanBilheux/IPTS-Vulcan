```python
import pandas as pd
import os
import numpy as np

from plotly.offline import plot, init_notebook_mode, iplot
init_notebook_mode()
import plotly.plotly as py
import plotly
import plotly.graph_objs as go

from ipywidgets import widgets

import gui_widgets
```

Checking that file exists

```python
%gui qt
```

```python
file_name = gui_widgets.gui_fname()
file_name = file_name[0]
```

Loading file using Pandas

```python
header = ['x', 'z', 'y', None, 'intensity']
df = pd.read_csv(file_name, sep='\t', names=header, skiprows=1)
df = df.reset_index()

# remove empty columns
try:
    del df[None]
except:
    pass

x = np.array(df['x'])
y = np.array(df['y'])
z = np.array(df['z'])
intensity = np.array(df['intensity'])
```

Display in 3D

```python
trace = plotly.graph_objs.Scatter3d( x=x,
                  y=y,
                  z=z,
                  mode = 'markers',
                  marker = dict(
                    sizemode = 'diameter',
                    sizeref=.05,
                    size= intensity,
                    colorscale = 'Viridis',
                    color = intensity,
                    colorbar = dict(title='Intensity'),
#                    line = dict(color='rgb(140,140,170)')
                    )
                  )
data = [trace]
layout = dict(height=800, width=800, title='Intensity')
fig = dict(data=data, layout=layout)
iplot(fig)
```
